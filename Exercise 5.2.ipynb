## Multiple plots

2.1) Draw a matrix contaning 4 plots using the `iris` dataset: in the upper left, plot the relationship between sepal length and sepal width over the whole data. The other three plots make the same comparison, but each divided in each of the species.

```r
# We import the library that we need 
library(datasets)

# Graph plot
par(mfrow=c(2, 2))
plot(iris$Sepal.Length, iris$Sepal.Width, xlab="Sepal Length", ylab="Sepal Width", col=iris$Species)
plot(iris$Sepal.Length[iris$Species=="setosa"], iris$Sepal.Width[iris$Species=="setosa"], xlab="Sepal Length", ylab="Sepal Width", col=1)
plot(iris$Sepal.Length[iris$Species=="versicolor"], iris$Sepal.Width[iris$Species=="versicolor"], xlab="Sepal Length", ylab="Sepal Width", col=2)
plot(iris$Sepal.Length[iris$Species=="virginica"], iris$Sepal.Width[iris$Species=="virginica"], xlab="Sepal Length", ylab="Sepal Width", col=3)
```

An alternative solution is

```r
par(mfrow=c(2,2))
species=levels(iris$Species)

plot(iris$Sepal.Length, iris$Sepal.Width, col=iris$Species)

for (i in 1:3) {
    plot(iris$Sepal.Length[iris$Species==species[i]], iris$Sepal.Width[iris$Species==species[i]], col=i)
}
```
