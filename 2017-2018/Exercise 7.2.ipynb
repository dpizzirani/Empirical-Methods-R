### Exercise 2.1 
Load the data frame `data.txt` from OLE
1. Using `sample` compute _5 different_ samples of size 50. For each of them, compute the sample mean and sample variance of the two variables in the data frame. Compare the results. Are they all the same?

```r
data <- read.table('data.txt', header=TRUE)
```

```r
z <- data.frame(replicate(5, sample(data$ht), 50))
z
```

```r
apply(z, 2, mean)
```

```r
apply(z, 2, var)
```

2. Use the sample size 100 to estimate the mean and variance of `ht`.

```r
x <- sample(x = data$ht, size = 100, replace = TRUE)
x
```

```r
mean(x)
var(x)
```

### Exercise 2.2
The column `g` from `data.txt` marks the gender of the sampled indivisial. Obrtain samples of 100 male and 100 female individuals from the data. 
1. What can you conclude of the difference in `ht` between genders? 
2. Make a boxplot from the sampled data (one box for each gender). Does the visualization confirm or refute your hypothesis from the last question?

```r
# The function data.frame is used to create a frame (a table unifying multiple values). In this case it
# contains data samples for males and females
y <- data.frame(sample(x = data$ht[data$g==1], size = 100), sample(x = data$ht[data$g==1], size = 100))
y
```

```r
apply(y, 2, mean)
apply(y, 2, var)
```

```r
boxplot(y)
```
