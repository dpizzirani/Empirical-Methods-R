## Multiple plots

2.1) Draw a matrix contaning 4 plots using the `iris` dataset: in the upper left, plot the relationship between sepal length and sepal width over the whole data. The other three plots make the same comparison, but each divided in each of the species.

```r
# We import the library that we need 
library(datasets)

# Graph plot
attach(mtcars)
par(mfrow=c(2, 2))
plot(iris$Sepal.Length, iris$Sepal.Width, xlab="Sepal Length", ylab="Sepal Width")
plot(iris$Sepal.Length[iris$Species=="setosa"], iris$Sepal.Width[iris$Species=="setosa"], xlab="Sepal Length", ylab="Sepal Width")
plot(iris$Sepal.Length[iris$Species=="versicolor"], iris$Sepal.Width[iris$Species=="versicolor"], xlab="Sepal Length", ylab="Sepal Width")
plot(iris$Sepal.Length[iris$Species=="virginica"], iris$Sepal.Width[iris$Species=="virginica"], xlab="Sepal Length", ylab="Sepal Width")
```
