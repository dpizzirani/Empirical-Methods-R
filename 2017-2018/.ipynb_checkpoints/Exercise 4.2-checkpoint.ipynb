## Exercise 3.2

Import the `datasets` library which contains a data set called `iris`. You can use the command `summary(iris)` to understand the variables inside thios data set. 

```r
library(datasets)
summary(iris)
```

From the `iris` dataset, plot the sepal length (in horizontal axis) vs. the sepal width (vertical axis), using colours and point shaped for the different species. Add a legend that distinguished the cases. 

First, we plot the sepal length and width compared

Then, we differentiate the different species by shape and by colour

```r
plot(iris$Sepal.Length[iris$Species=="setosa"], iris$Sepal.Width[iris$Species=="setosa"], xlab="Sepal Length",ylab="Sepal Width")
points(iris$Sepal.Length[iris$Species=="versicolor"], iris$Sepal.Width[iris$Species=="versicolor"], col=2, pch=2)
points(iris$Sepal.Length[iris$Species=="virginica"], iris$Sepal.Width[iris$Species=="virginica"], col=3, pch=3)
```

Finally, we add the legend

```r
plot(iris$Sepal.Length[iris$Species=="setosa"], iris$Sepal.Width[iris$Species=="setosa"], xlab="Sepal Length",ylab="Sepal Width")
points(iris$Sepal.Length[iris$Species=="versicolor"], iris$Sepal.Width[iris$Species=="versicolor"], col=2, pch=2)
points(iris$Sepal.Length[iris$Species=="virginica"], iris$Sepal.Width[iris$Species=="virginica"], col=3, pch=3)
legend(4.3, 4.3, c("Setosa","Versicolor","Virginica"), col=c(1,2,3), pch=c(1,2,3))
```

Another shorter solution can be

```r
plot(iris$Sepal.Width, iris$Sepal.Length, col=iris$Species, xlab="Sepal Width", ylab="Sepal Length")
legend("topright", legend=levels(iris$Species), pch=1, col=1:3)
```
