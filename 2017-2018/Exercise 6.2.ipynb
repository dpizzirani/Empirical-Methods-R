### Exercise 3
1. Create a scatter plot of the variable `MPG.city` depending on `Weight`
2. Color the points corresponding to the variabe `Cylinders`
3. Use for the different number of `Origin` a different symbol
4. Add a legend to describe the number of gears

```r
# Type indicates no plot, yaxt suppresses y axis
plot(MPG.city ~ Weight, data=Cars93, type='n')
points(MPG.city ~ Weight, data=Cars93, pch=as.integer(Cars93$Origin), cex=3, col=as.integer(Cars93$Origin), yaxt="n")
legend("topright", pch=as.integer(Cars93$Origin), legend=c("non-USA", "USA"))
axis(1, at=2000, label="twohun")
```

### Exercise 4
Visualize the counts of `Type` using a piechart(`pie()`) and a barchart (`barplot()`). In whcih plot is the data easier to understand?

```r
par(mfrow=c(1,2))
pie(table(Cars93$Type))
barplot(table(Cars93$Type), las=3)
```

Another solution which just some graphical changes

```r
par(mfrow=c(1,3), cex.axis=1.7, cex.lab=1.7)
pie(table(Cars93$Type), cex=1.7)
barplot(table(Cars93$Type), ylab="count", las=3)
barplot(table(Cars93$Type), horiz=TRUE, xlab="count", las=1)
```
