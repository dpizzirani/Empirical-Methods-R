Access to a value of the table

```r
table$x
```

__1) Compute the range of the variable__ `x`

```r
range(table$x)
```

__2) Compute the mean value of__ `x`

```r
mean(table$x)
```

__3) Round the product of__ `x` __and__ `y` __using the function__ `round()`

```r
round(table$x * table$y)
```

## Exercise 2.2
1. What is the probability that _x ≤ 6_?

```r
# length(): number of elements
# nrow(): number of rows
length(which(table$x<=6))/nrow(table)
```

2. What is the probability that _x ≤ 6_ given that _y ≤ 6_?

```r
length(which(table$x<=6 & table$y<=6))/length(which(table$y<=6))
```

3. What is the probability that _x ≥ 5_ given that _x > y_?

```r
length(which(table$x>=5 & table$y< table$x))/length(which(table$y< table$x))
```

4. What is more likely, _x ≤ 3.6_, or _y ≤ 4.9_? Could you answer this question just looking at the summary of the table?

To answer this question we may observe that 4.9 is the median of y, but 3.6 is much larger than the median of x; so the former is more likely (even without computing the specific values).
