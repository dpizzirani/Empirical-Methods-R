# Basic Plotting
You are going to learn how to plot using the `ggplot` library of R. Download and read the handout on this topic.
Read the table in the `data.txt` file in OLE.

## Exercise 2.1

```r
library(ggplot2)
table <- read.table("data.txt")
```

1. Draw a histogram with 10 breaks for the variable x in the data.

```r
ggplot(table, aes(x)) + geom_histogram()
```

2. Include, below the histogram a strip chart representing the actual values, in groups
of width 0.1.

```r
ggplot(table,aes(x)) + geom_histogram() + geom_rug()
```

3. Draw the histograms for x and y in a single plot. Try to make it such that the two
histograms are clearly visible.

```r
ggplot(table,aes(x))+geom histogram()+geom rug()+geom histogram(aes(y),fill=2,alpha=0.5)
```
